# Vectorización

Consiste en transformar el texto, en concreto los tokens, en vectores de números, que serán la entrada de los algoritmos de clasificación. 



```
from sklearn.feature_extraction.text import CountVectorizer
# list of text documents
texts = ["The apple is rotten",
         "The children were playing",]
# create the transform
vectorizer = CountVectorizer()
# tokenize and build vocab
vectorizer.fit(texts)
# summarize
print(vectorizer.vocabulary_)
print('Tamaño del vocabulario:',len(vectorizer.vocabulary_))

# encode document

```

```
for text in texts:
  vector = vectorizer.transform([text])
  # summarize encoded vector
  #print(vector.shape)
  #print(type(vectors))
  print(text,vector.toarray())
```

## Tf-idf

```
from sklearn.feature_extraction.text import TfidfVectorizer
# list of text documents
texts = ["The apple is rotten",
         "The child is playing",
         "The zoo"]
# create the transform
vectorizer = TfidfVectorizer()
# tokenize and build vocab
vectorizer.fit(texts)
# summarize
print(vectorizer.vocabulary_)
print(vectorizer.idf_)
```

```
for text in texts:
  vector = vectorizer.transform([text])
  # summarize encoded vector
  #print(vector.shape)
  #print(type(vectors))
  print(text,vector.toarray())
```
